crie funções:

def salva_bd(df, nome_tabela): salva dataframe df na tabela nome_tabela.


def carrega_bd(nome_tabela): carrega tabela nome_tabela num dataframe.

```python
def salva_bd(df,nome_tabela,nome_db):
    ''' essa função é utilizada pra criar um banco de dados na conexão 
    '''
    conn = sqlite3.connect('coderhouse.db')
    df.to_sql(nome_tabela,conn,if_exists='replace',index=False) # salva o dataframe no banco
    conn.close()

    return True
```

```python
def carrega_bd(nome_tabela):
    '''essa função é utilizada pra criar um dataframe a partir da tabela do banco de dados '''
    conn = sqlite3.connect('coderhouse.db')
    query = f'SELECT * FROM {nome_tabela}'
    df = pd.read_sql(query,conn)
    conn.close()

    return df
```

```python
def tabelas_bd():
    '''
        Retorna um dataframe com as tabelas do banco de dados.
    '''
    conn = sqlite3.connect('coderhouse.db')

    # Executar uma consulta que retorna as informações do esquema do banco de dados
    query = "SELECT name FROM sqlite_master WHERE type='table'"
    schema = pd.read_sql_query(query, conn)

    conn.close()

    return schema
```
