## Estruturas de Repetição

Outra estrutura fundamental de algoritmos, enquanto a condição for verdadeira, essa estrutura irá se repetir. temos como estruturas classicas de repetição loops while e loops for.

image.png

### While

é Uma estruura de repetição que irá se repetida até que o valor da condição seja verdadeiro.

```python
soma = 1
while soma < 12:
    soma = soma + 2
    print(soma)
```

garantir que nunca faça um loop como esse:

```python
soma = 1
while soma != 12: # CUIDADO!!!
    soma = soma + 2
    print(soma)
```

Exercício 2 - Crie um programa em Python que solicita ao usuário uma senha numérica de 4 dígitos. O programa deve repetir essa solicitação até que o usuário informe a senha correta (1234).

```python
senha = input("Digite a senha numérica de 4 dígitos: ")
while senha != "1234":
    senha = input("Senha incorreta. Digite a senha numérica de 4 dígitos novamente: ")
print("Senha correta!")
```

### For

Estrutura de repetição usada pra quando sabemos exatamente o número de vezes que a estrutura precisa ser repetida.Em Python definimos o for para um intervalo, que pode ser igual ao tamanho de uma lista/tupla ou mesmo pré definido com a função built-in range().

```python
for i in [3,5,7,9,13]:
    print(i)
```

assim como passamos explicitamente a lista, podemos passar como variável salva em memória.

```python
lista_exemplo = [3,5,7,9,13]
for i in lista_exemplo:
    print(i)
```

Calculando as potências de 10

```python
pot_10 = [1,10,100,1000,10000]
count = 0
for potencia in pot_10:
  print(f'10^{count}={potencia}')
  count += 1
```

Exercício 3 - Escreva um programa em Python que solicita ao usuário uma frase e, em seguida, exibe cada palavra e seu número de caracteres.

```python
frase = input("Digite uma frase: ")

palavras = frase.split()

for palavra in palavras:
    tamanho = len(palavra)
    print(palavra, tamanho)
```

Podemos interar nos caracteres de uma string:

```python
for c in "Ola":
    print(c)
```

podemos criar loops for dentro de loops for.

```python
for i in [1,2]:
    for j in [3,4]:
        print(i,j)
```

### Utilizando a função Built-in Range()
Essa função é utilizada pra criar intervalos de interação para sere utilizados em loops for.

```python
for i in range(4):
    print(i)
```

```python
for i in range(1,5):
    print(i)
```

```python
for i in range(20,60,10):
    print(i)
```

Exercício 4 - Escreva um programa que dado um valor n calcula o fatorial de n.

```python
n = int(input("Digite um número inteiro: "))
fatorial = 1

for i in range(1, n+1):
    fatorial *= i

print("O fatorial de", n, "é", fatorial)
```
