### Operadores Aritméticos

são as operações aritméticas básicas que conhecemos da matemática.

```python
soma = x + y
sub  = x - y
mul  = x*y
div  = x/y
print(soma,sub,mul,div)
```

```python
print(x != y)
print(x > y)
print(x < y)
```

```python
print(x > y*2)
print(x >= y*2)
```

## Estruturas de Dados

de modo geral guardam o que chamamos de coleções de dados,  no python temos Listas, Tuplas, Dicionários e set como principais estruturas de dados

### Listas
pode ser alterada (mutável), é **ordenada**, e pode conter numeros arbitrários ou qualquer tipo de dado, inclusive outras listas (multinível).

```python
minha_lista = [1, 2.2, "três", 4.0]
minha_lista[2]
```

 - Listas podem ser multiníveis, ou seja listas incluirem outras listas

```python
minha_lista2 = [5, 6, [7, 8]]
minha_lista2
```

 - podemos adicionar novos itens o método .append() que irá incluir no final da lista um novo item.

```python
minha_lista2.append('nove')
minha_lista2
```

 - podemos remover itens usando o método .pop(). Se não passarmos um item do index ela exclui o final.

```python
minha_lista2.pop()
print(minha_lista2)

minha_lista2.pop(0)
print(minha_lista2)
```

 -  Podemos concatenar listas

```python
lista_completa = minha_lista + minha_lista2
lista_completa
```

 - Para ordernarmos podemos utilizar o método .sort() que vai arrumar de forma ascendente. Para forma Descendente na ordenação é só passar o argumento .sort(reverse=True)

```python
#Ascendente
lista_ordernar = [5,2,1,3,4]
lista_ordernar.sort()
print(lista_ordernar)
#Descendente
lista_ordernar.sort(reverse=True)
print(lista_ordernar)
```

#### Dicionários
É uma estrutura de dados mutável, não ordenada, que é mapeada por chaves e valores

```python
meu_dict = {"chave1": "valor1", "chave2": 2, "chave3": [1, 2, 3]}
meu_dict
```

 -  Para adicionar um novo item a um dicionário, é só passar a nova chave da forma dicionario["nova_chave"] = novo_valor

```python
joao = {"nome":"Joao Silva",
        "idade":56}

joao["altura"] = 1.81

joao
```

#### Tuplas

Tuplas são imutáveis, ou seja, não podemos adicionar novos itens a tupla. semelhante a lista, possui valores ordenados por índice.

```python
minha_tupla = (1, "dois", 3.0)
minha_lista = [1, "dois", 3.0]
```

```python
minha_lista[1] = 2
minha_lista
```

```python
minha_tupla[1] = 2
minha_tupla
```

#### Sets

Sets são um tipo de estrurura ou coleção que é mutável e não ordenada. Itens em sets não podem ser duplicados, então apenas 1 item para cada elemento é permitido. exe: não podemos ter duas entradas "maça" em um set.

para acessarmos itens em um set, apenas com loops do tipo for. então isso será abordado novamente na aula 3.

```python
meu_set = {1,2,2,3,4,5,2}
meu_set
```

geralmente usamos itens de um set para fazer casting de listas e tuplas.

```python
agora_lista = list(meu_set)
agora_tupla = tuple(meu_set)
print(meu_set)
print(agora_lista)
print(agora_tupla)
```

#### string.upper()
colocamos todos os caracteres em upper case -  letras maísculas.



```python
texto = "Hello World!"
texto_maiusculo = texto.upper()
print(texto_maiusculo)
```

#### string.lower()

```python
texto = "Hello World!"
texto_minusculo = texto.lower()
print(texto_minusculo)
```
