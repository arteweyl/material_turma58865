## Via SSH
Secure Shell ainda é a linha numero um para configuração do servidor git remoto, aqui o github, então é boa prática sempre setar via ssh, porém ele dá bem mais trabalho para primeira configuração.

### Criando a chave de autenticação ssh

Aqui no Vs code vá no simbolo de mais e em git bash

Faça o comando no git bash:

```bash
 ssh-keygen -t ed25519 -C "seu@email.com"
```

Ele vai pedir pra criar uma senha, não precisa por nenhuma.
ele vai salvar um par publico e privado e vai salvar em um endereço assim:
c:/Users/YOU/.ssh/id_ed25519

### adicionando a chave ao agente ssh
Abra o Powershell com permissões de admin e iniciar o serviço

Rode os comandos que vão iniciar o serviço ssh na sua maquina.


```bash
Get-Service -Name ssh-agent | Set-Service -StartupType Manual
Start-Service ssh-agent
```

voltando ao vs code, da mesma forma que iniciou o git bash, inicie uma nova aba do powershell sem permissão de admin( é o padrão do vs code)

execute o comando que de fato adiciona a chave criada anteriormente no agente ssh:

```bash
ssh-add c:Users/lcwey/.ssh/id_ed25519
```
