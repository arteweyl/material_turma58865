```python
import schedule
import time
```

```python
def minha_tarefa():
    print("executando minha tarefa")
    schedule.every(1).seconds.do(minha_tarefa)

while True:
    minha_tarefa()
    schedule.run_pending()
    time.sleep(1)
```
