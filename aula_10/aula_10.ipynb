```python
data1 = {'produto':['relogio','anel','brinco','camisa'],
                    'loja':['Zimer','Piter','Piter','Loyyn'],
                    'valor':[100, 70, 20, 120]}
df1 = pd.DataFrame(data1) 
df1
```

```python
data2 = {'produto':['gravata','colar','bolsa'],
                    'loja':['Loyyn','Piter','Piter'],
                    'valor':[60, 120, 300]}

df2 = pd.DataFrame(data2)
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
data3 = {'cidade':['Loyyn','Piter','Zimer'],
                    'loja':['New Romy','Piter','Golgo']}

df3 = pd.DataFrame(data3)
df3
```

```python
pd.concat([df1,df2,df3],ignore_index=False)
```

```python
pd.concat([df1,df3],axis=1)
```

### Merge (Join)

```python
df1.merge(df3,on='loja',how='inner')
```

```python
df1.merge(df3,on='loja',how='right')
```

```python
df1.merge(df3,on = 'loja',how='outer')
```

#### Exercício
Aplique as quatro formas de join estudadas nos seguintes dataframes:

df1 = pd.DataFrame({'A': [4, 2, 1, 3], 'B': ['a', 'b', 'c', 'b']})

df2 = pd.DataFrame({'A': [2, 3, 5], 'C': [10, 20, 30]})

Sobre o resultado do outer join anterior, ordene a saída pela coluna A (ajustar índices).

```python
df1 = pd.DataFrame({'A': [4, 2, 1, 3], 'B': ['a', 'b', 'c', 'b']})
df1
```

```python
df2 = pd.DataFrame({'A': [2, 3, 5], 'C': [10, 20, 30]})
df2
```

```python
df1.merge(df2,on='A',how='left')
```

```python
df1.merge(df2,on='A',how='right')
```

```python
df1.merge(df2,on='A',how='inner')
```

```python
df1.merge(df2,on='A',how='outer')
```

```python
outer = df1.merge(df2,on='A',how='outer')
outer = outer.sort_values('A')
outer = outer.reset_index(drop=True)
outer
```

Concatenar df1 e df2 em formato de linhas e depois colunas (ajustar índices)
