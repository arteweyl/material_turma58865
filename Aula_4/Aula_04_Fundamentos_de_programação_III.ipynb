## Funções

Podemos definir valores padrões para os parâmetros de uma função, o que significa que, se um argumento não for fornecido quando a função for chamada, ele usará o valor padrão definido em vez disso.


```python
def mul(a,b):
    x = a*b
    return x
mul(4,5)
```

```python
print(mul(4,5))
print(mul(-2,10))
print(mul(4,mul(2,5)))
```

```python
def ordena(a,b):
    if a > b:
        return a, b
    return b, a

print(ordena(5,3))
print(ordena(3,5))
```

#### Exércicio
 Converta o programa de cálculo de fatorial da aula anterior para uma função. A função recebe como argumento o n e retorna o fatorial de n.

```python
def calc_fatorial(n):

    fatorial = 1

    for i in range(1, n+1):
        fatorial *= i

    return fatorial
calc_fatorial(5)
```

## Lambda (Função Anônima)

Linguagens de programação geralmente possuem recursos de funções anonimas, que são funções que são armazenadas ddiretamente em variáveis em memória, sendo chamadas toda vez que chamamos a variável.

```python
quadrado = lambda x: x ** 2
quadrado(4)
```

```python
eh_par = lambda x: (x % 2) == 0
eh_par(4)
```

## Map

map é uma função Built-in que recebe uma função e uma estrutura de dados como argumentos, e aplica a função a cada elemento da estrutura, retornando um dado do tipo map, que mapeou cada item da estrutura anterior. Para tornar funcional, só criar um cast da variável armazenada pelo map, podendo ser list(), tuple() ou set().


```python
frutas = ['Banana', 'Pera', 'Tomate']
qtd_letras_frutas = map(len, frutas)
print(list(qtd_letras_frutas))
```

```python
tamanho_nome = map(len,["joao",'maria','jose'])
print(list(tamanho_nome))
```

```python
lista_quadrados = list(map(quadrado, (3,4,5)))
lista_quadrados
```

```python
def substitui_a_por_x(palavra):
    return palavra.replace('a', 'x')


novas_frutas = {'banana', 'maçã', 'laranja', 'abacaxi'}


resultado = map(substitui_a_por_x, novas_frutas)
print(resultado)

nova_lista = list(resultado)

print(nova_lista)
```

Exercício  - Crie uma função primeiro_nome que recebe uma string nome completo e retorna somente o primeiro nome. Em seguida, use a função map para aplicar a função a uma lista de strings:nomes = ['João Carlos', 'Maria Eduarda', 'Ana Rute', 'Ana Clara']

```python
def primeiro_nome(nome_completo):
    nome_split = nome_completo.split()
    return nome_split[0]

nomes = ['João Carlos', 'Maria Eduarda', 'Ana Rute', 'Ana Clara']
so_primeiro_nome = list(map(primeiro_nome, nomes))

print(so_primeiro_nome)
```

Exercício - escreva uma função que receba uma data no formato dd/mm/aaaa e retorne o dia da semana correspondente (Monday, Tuesday, etc.).


```python
from datetime import datetime

def dia_da_semana(data):
    # converte a string para datetime
    data = datetime.strptime(data, '%d/%m/%Y')
    # formata a data para obter o nome do dia da semana
    nome_dia_da_semana = data.strftime('%A')

    return nome_dia_da_semana

# exemplo de uso da função
data = '10/04/2023'
print(dia_da_semana(data))
```

```python
#por isso definam nomes de funções coma  funcionalidade certa, é boa prática.

def celsius_para_fahrenheit(celsius):
    """
    Converte uma temperatura em
    graus Celsius para Fahrenheit.
    """
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
```
