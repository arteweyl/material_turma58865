```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fastfood = pd.read_excel('../datasets/fastfood.xlsx')
fastfood.head(1)
```

```python
fastfood_agg = fastfood.groupby('restaurant').mean(numeric_only=True).reset_index()
fastfood_agg.head()
```

```python
plt.bar(fastfood_agg['restaurant'],fastfood_agg['calories'],color='blue')
plt.xticks(rotation=45)
plt.title('Mdia de calorias por Restaurante')
plt.xlabel('Restaurante')
plt.ylabel('Mdia de Calorias')
plt.show()
```

```python
plt.hist(fastfood['calories'],bins=8)
plt.show()
```

```python
sns.displot(fastfood['calories'],kde=True)
plt.show()
```

```python
sns.scatterplot(x='calories', y='sugar',data=fastfood)
plt.show()
```

```python
sns.boxplot(x='restaurant',y='calories',data=fastfood)
plt.xticks(rotation=45)
plt.show()
```
