### Uso

#### Array unidimensional

```python
arr = np.array([1,2,3,4])
print(arr)
```

#### Array bidimensional(Matriz)

```python
matriz = np.array([[1.0,2],[3,4]])
print(matriz)
```

#### Array tri ou ndimensional(tensor)

```python
tensor = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(tensor)
```

#### Atributos de array

```python
print(matriz.ndimension) #quantidade de dimensões do array
print(matriz.shape) #formato do array, quantidade de linhas/colunas
print(tensor.shape) #mesma coisa, só que utilizado em tensor que possui 3 ou mais dimensões além de linhas e colunas.
```

```python
print(matriz.dtype) #tipo de dados contidos no array
print(tensor.dtype)
```

#### Operações de agregação com arrays

```python
meu_array2 = np.array([1,2,3])
```

```python
print(f"Soma: {meu_array2.sum()}")
print(f"Max: {meu_array2.max()}")
print(f"Min: {meu_array2.min()}")
print(f"média: {meu_array2.mean()}")
print(f"Desvio Padrão: {meu_array2.std()}")
print(f"Variancia: {meu_array2.var()}")
```
